# Step 5: Questions to answer about the data

```python
 #Q1: What are the locations with the fastest roller coasters (minimum of 10 coasters)?
ax = df.query('Location != "Other"') \
    .groupby('Location')['Speed_mph'] \
    .agg(['mean', 'count']) \
    .query('count >=10') \
    .sort_values(by='mean', ascending = False)['mean'] \
    .plot(kind = 'barh', title = 'Top Locations with fastest roller coasters (minimum of 10 coasters)', xlabel='Average Speed (mph)', color='darkviolet')
ax.bar_label(ax.containers[0], label_type = 'edge', fmt='%.1f')
plt.show()
```

```python
# Q2:  Which locations have the most number of operational coasters (still operating)?
ax = df.query('Status =="Operating" and Location != "Other"') \
    .groupby(['Location'])['Coaster_Name'] \
    .agg(['count']) \
    .sort_values(by = 'count',ascending = False) \
    .head(10) \
    .plot(kind='barh', title = 'Top locations having the most number of operational coasters (still operating)', 
          xlabel='Number of Operational Coasters', color ='navy')
ax.bar_label(ax.containers[0], label_type = 'edge')
ax.invert_yaxis()
plt.legend().remove()
plt.show()
    
```

```python
# Q3: Top Manufacturers with most number of roller coasters
ax = df['Manufacturer'].value_counts() \
    .head(5) \
    .plot(kind = 'bar', title='Top 5 Manufacturers with most number of roller coasters', ylabel ='Number of Coasters', color='limegreen', figsize=(8,7))
ax.bar_label(ax.containers[0], label_type='edge')
plt.xticks(rotation=0)
plt.show()
```
